```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "../data/labels/patient_label.csv"  
df = pd.read_csv(csv_path)

label_counts = df["histo_label"].value_counts().sort_index()

plt.figure(figsize=(5, 4))
label_counts.plot(kind="bar", color=["skyblue", "salmon"])
plt.xticks(ticks=[0, 1], labels=["Benign (0)", "Malignant (1)"], rotation=0)
plt.ylabel("Count")
plt.title("Count of Benign vs Malignant Cases")
plt.tight_layout()
plt.show()
```
